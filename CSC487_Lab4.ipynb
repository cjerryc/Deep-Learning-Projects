```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
import keras

# Load the CIFAR10 dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
# Preprocess the data: scale the images and convert the labels to categorical format.
# Scale images down
train_images, test_images = train_images / 255.0, test_images / 255.0

# Randomly shuffle the training data and labels, while preserving label-image correlation
train_images, train_labels = shuffle(train_images, train_labels)
# Separate a portion of images and labels from training set to be validation data
val_size = train_images.shape[0] // 10       # Separate out a validation set size of 1/10 samples (5,000 in this case)

val_images = train_images[:val_size]
val_labels = train_labels[:val_size]
train_images = train_images[val_size:]
train_labels = train_labels[val_size:]

```

```python
# Build a CNN model with TensorFlow.
model = models.Sequential()
model.add(layers.Conv2D(64, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(layers.MaxPooling2D((3, 3)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))

# # Add dense layers on top
# Flatten multi-dimensional input into 1-D
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(10))

# Train the model on the training data.
# Loss = sparse categorical entropy to create a probability distribution across 10 possible classes (final output)
model.compile(optimizer=keras.optimizers.Adam(lr=0.001),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
convHistory = model.fit(train_images, train_labels, epochs=10, validation_data=(val_images, val_labels))


# Evaluate the model on the test data.
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print('\nTest accuracy:', test_acc)
```
